# **Labour Market Trends and Skill Shortages Analysis**
***Objective***: Clean the dataset downloaded from Kaggle


***Dataset***: https://www.kaggle.com/datasets/lukebarousse/data-analyst-job-postings-google-search

```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Projects/Data Visualisation/Jobs.csv")
print(df.shape)
print(df.columns,'\n')
df.head()
```

```python
# Drop unwanted columns
columns_to_remove = ['Unnamed: 0', 'index', 'via', 'thumbnail', 'commute_time', 'description_tokens', 'job_id']
df = df.drop(columns=columns_to_remove, errors='ignore')
```

```python
# Rename columns to Title Case
df.columns = [col.replace('_', ' ').title() for col in df.columns]
print(df.columns,'\n')

# Display updated dataframe
df.head()
```

```python
# List unique values and the count for each column
for col in df.columns:
    unique_values = df[col].dropna().unique()  # Remove NaN values and get unique values
    num_unique = len(unique_values)  # Count the number of unique values
    print(f"Column: {col}")
    print(f"Number of unique values: {num_unique}")
    print(unique_values)
    print("-" * 50)
```

```python
import pandas as pd
import ast  # To safely convert string representation of list to actual list

# Convert 'Extensions' column from string to list
df['Extensions'] = df['Extensions'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to extract values based on patterns
def extract_extension_value(ext_list, keyword):
    """Extracts first matching value from the list based on keyword"""
    return next((item for item in ext_list if keyword in item.lower()), None)

# Extract new columns from 'Extensions'
df['Time Posted'] = df['Extensions'].apply(lambda x: next((i for i in x if 'hour' in i or 'day' in i), None))
df['Salary Range'] = df['Extensions'].apply(lambda x: next((i for i in x if '–' in i or '$' in i or '£' in i or 'an hour' in i or 'a year' in i), None))
df['Job Type'] = df['Extensions'].apply(lambda x: extract_extension_value(x, 'full-time') or extract_extension_value(x, 'part-time'))

# Extract remaining values as 'Benefits'
df['Benefits'] = df.apply(lambda row: ', '.join([i for i in row['Extensions'] if i not in [row['Time Posted'], row['Salary Range'], row['Job Type']]]), axis=1)

# Drop the original 'Extensions' column
df.drop(columns=['Extensions'], inplace=True)

# Display updated dataframe
df.head()
```

```python
df.columns.tolist()
```

```python
# Drop unwanted columns
columns_to_remove = ['Job Type', 'Salary', 'Time Posted']
df = df.drop(columns=columns_to_remove, errors='ignore')
df.columns.tolist()
```

```python
# Assuming your data is stored in a DataFrame called df
df['Work From Home'] = df['Work From Home'].replace({True: 'Yes', 'NAN': 'No'})

# To handle actual NaN values in the column, use pd.NA or np.nan
df['Work From Home'] = df['Work From Home'].fillna('No')
```
